```python
import numpy as np
from transformers import pipeline
import matplotlib.pyplot as plt
```

```python
zsc = pipeline('zero-shot-classification', model='facebook/bart-large-mnli')
```

```python
result = zsc(
    sequences="Can I also get a screen protector and a protective case with my purchase?",
    candidate_labels=['Product Inquiry', 'Price Inquiry', 'Promotion & Discount Inquiry', 'Accessory Inquiry', 'Purchase'],
    multi_label=False
)
```

```python
plt.bar(result['labels'], result['scores'])
plt.yticks(list(np.arange(0,1,0.1)))
plt.xticks(rotation=90)
plt.show()
```
